Creating 3 new Parameters in the dataset , to judge the past interactions of the user at the last three stalls they visited.
This is a part of using the Lead Capturing of the user
These parameters are : 
* staff engagement
* informativeness
* overall experience



```python
import numpy as np
df['param1'] = None
df['param2'] = None
df['param3'] = None
for idx , rows in df.iterrows():
    if rows['pastinter'] == 'yes':
        df.at[idx, 'param1'] = [np.random.randint(1, 10),np.random.randint(1, 10),np.random.randint(1, 10)]
        df.at[idx, 'param2'] = [np.random.randint(1, 10),np.random.randint(1, 10),np.random.randint(1, 10)]
        df.at[idx, 'param3'] = [np.random.randint(1, 10),np.random.randint(1, 10),np.random.randint(1, 10)]
    else:
        df.at[idx, 'param1'] = [np.random.randint(1, 10),np.random.randint(1, 10),np.random.randint(1, 10)]
        df.at[idx, 'param2'] = [np.random.randint(1, 10),np.random.randint(1, 10),np.random.randint(1, 10)]
        df.at[idx, 'param3'] = [np.random.randint(1, 10),np.random.randint(1, 10),np.random.randint(1, 10)]
```
