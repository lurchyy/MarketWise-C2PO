```python
vis = pd.read_csv('visitors.csv')
exhib = pd.read_csv('exhibitors (1).csv')
```

```python
empty_col = []
for col in vis.columns:
    if vis[col].isnull().all() or vis[col].isna().all():
        print(col)
        empty_col.append(col)
```

```python
vis = vis.drop(columns=empty_col)
vis.head()
```

```python
import matplotlib.pyplot as plt
import numpy as np

index = np.arange(len(vis.columns))   # label locations
present = len(vis) - vis.isnull().sum()
missing = vis.isnull().sum()



fig, ax = plt.subplots()

bar1 = ax.bar(index, present,
              label='Present',
              color='b')

bar2 = ax.bar(index,
              missing,
              label='Missing',
              color='orange')

ax.set_xlabel('Categories')
ax.set_ylabel('Count')
ax.set_title('Count by category and presence')
ax.set_xticks(index)
ax.set_xticklabels(vis.columns, rotation=90)
ax.legend()

plt.show()
```

```python
df = pd.read_csv('vis_fin.csv')
```

```python
empty_col = []
for col in df.columns:
    if df[col].isnull().all() or df[col].isna().all():
        print(col)
        empty_col.append(col)
df = df.drop(columns=empty_col)
```

```python
exhib.head()
```

```python
import random

# Get 3 random values from company_profile column in exhib DataFrame
random_values = exhib['company_name'].sample(n=207).tolist()

# Create pastinter column in df DataFrame
df['pastinter'] = ''

# Populate pastinter column with random values
for index, row in df.iterrows():
    random_values_selected = random.sample(random_values, 3)
    df.at[index, 'pastinter'] = random_values_selected

df.head()
```
