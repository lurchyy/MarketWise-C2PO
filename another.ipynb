```python

def process_text(text):
    """Process text function.
    Input:
        text: a string containing the text
    Output:
        processed_text: a list of words containing the processed text

    """
    stemmer = PorterStemmer()
    stopwords_english = stopwords.words('english')

    # remove stock market tickers like $GE
    text = re.sub(r'\$\w*', '', text)
    # remove old style retweet text "RT"
    text = re.sub(r'^RT[\s]+', '', text)
    # remove hyperlinks    
    text = re.sub(r'https?://[^\s\n\r]+', '', text)
    # remove hashtags
    # only removing the hash # sign from the word
    text = re.sub(r'#', '', text)
    # tokenize text
    tokenizer = TweetTokenizer(preserve_case=False, strip_handles=True,
                               reduce_len=True)
    text_tokens = tokenizer.tokenize(text)

    processed_text = ""
    for word in text_tokens:
        if (word not in stopwords_english and  # remove stopwords
                word not in string.punctuation):  # remove punctuation
            # processed_text.append(word)
            stem_word = stemmer.stem(word)  # stemming word
            processed_text+= " " + stem_word
    
    return processed_text

nltk.download('stopwords')  
```

```python
df_exb['text'] = df_exb['company_profile'].to_string()
```

```python
df_exb["preprocessed_text"] = df_exb["text"].apply(process_text)
df_exb["preprocessed_text"] = df_exb["preprocessed_text"].apply(lambda x: ' '.join(x)) #joins the list of processed words into a single string with space.
```

```python
import word2vec
```

```python
from gensim.models import Word2Vec

# Train word2vec model
sentences = [text.split() for text in df_exb['preprocessed_text']]
model = Word2Vec(sentences, min_count=1)

# Vectorize the processed text
vectorized_text = []
for text in sentences:
    vector = []
    for word in text:
        if word in model.wv:
            vector.append(model.wv[word])
    vectorized_text.append(vector)

# Assign the vectorized text back to the dataframe
df_exb['vectorized_text'] = vectorized_text

```

```python
def get_word2vec_vector(sentence):
    # Preprocess the sentence
    processed_sentence = process_text(sentence)
    
    # Initialize an empty vector
    vector = np.zeros(100)  # Assuming the Word2Vec model has 100-dimensional vectors
    
    # Calculate the vector representation
    for word in processed_sentence:
        if word in model.wv:
            vector += model.wv[word]
    
    return vector
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5)
```

```python
vectorized_text = df_exb['vectorized_text'].to_list()
reshaped_vectorized_text = [item for sublist in vectorized_text for item in sublist]
kmeans.fit(reshaped_vectorized_text)


```

```python
def find_nearest_neighbors(sentence, k=5):
    # Preprocess the sentence
    processed_sentence = process_text(sentence)
    
    # Vectorize the processed sentence
    vector = get_word2vec_vector(processed_sentence)
    
    # Predict the cluster for the vectorized sentence
    cluster = kmeans.predict([vector])[0]
    
    # Get the indices of the sentences in the same cluster
    cluster_indices = np.where(kmeans.labels_ == cluster)[0]
    
    # Check if cluster_indices is empty
    if len(cluster_indices) == 0:
        return []
    
    # Calculate the distances between the vectorized sentence and the sentences in the same cluster
    distances = [np.linalg.norm(vector - vectorized_text[i]) for i in cluster_indices]
    
    # Check if distances is empty
    if len(distances) == 0:
        return []
    
    # Sort the distances and get the indices of the nearest neighbors
    nearest_indices = np.argsort(distances)[:k]
    
    # Check if nearest_indices is empty
    if len(nearest_indices) == 0:
        return []
    
    # Get the nearest neighbors' sentences
    nearest_neighbors = [df_exb['text'].iloc[cluster_indices[i]] for i in nearest_indices if i < len(cluster_indices)]
    
    return nearest_neighbors

print(find_nearest_neighbors("I am a software engineer with experience in Python and Java.", k=5))
```
